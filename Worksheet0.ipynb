```python
def knapsack(weights, values, capacity):
    n = len(weights)

    # Step 1: Create DP table
    # dp[i][w] = max value using first i items and capacity w
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    # Step 2: Fill the table
    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i-1] <= w:
                # Include the item or skip it
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                # Cannot include the item, skip it
                dp[i][w] = dp[i-1][w]


    return dp[n][capacity]

weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7

print(knapsack(weights, values, capacity))
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)

    dp = [[0] * (m + 1) for _ in range(n+1)]

    for i in range(1, n+1):
        for j in range(1, m+1):

            if s1[i-1] == s2[j-1]:
                dp[i][j] = 1 + dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[n][m]

print(longest_common_subsequence("abcde", "abc"))
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic
    programming (tabulation).

    Parameters:
        coins (list of int): Available coin denominations.
        amount (int): Target amount to make change for.

    Returns:
        int: Minimum number of coins needed. Returns -1 if amount cannot be made.

    Example:
    >>> min_coins([1, 2, 5], 11)
    3
    >>> min_coins([2], 3)
    -1
    """
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)


    return dp[amount] if dp[amount] != float('inf') else -1

print(min_coins([1,2,5],11))
print(min_coins([5,5,1],3))
```

```python
def calculate_directory_size(directory):
    """
    This function recursively calculates the total size of a directory

    Args:
        directory (dict): Dictionary representing files and subdirectories

    Returns:
        int: Total size of directory including all nested files/sudirectories.
    """

    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

total_size = calculate_directory_size(directory_structure)
print("Total directory size:", total_size, "KB")
```

```python
def generate_permutations(s):
    """
    This function takes a string and recursively generates all unique permutations of the string.

    Args:
        s (str): Input String

    Returns:
        list: A list of unique permutations of the string.
    """

    if len(s) <= 1:
        return s

    permutations = []

    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]
        for p in generate_permutations(remaining):
            permutations.append(char + p)

    return list(set(permutations))
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def sum_nested_list(nested_list):
    """
    Calculate the sum of all numbers in a nested list.
    This function takes a list that may contain integers and other nested lists.
    It recursively traverses the list and sums all the integers, no matter how deeply
    nested they are.

    Args:
        nested_list (list): A list that may contain integers or other lists of integers.

    Returns:
        int: The total sum of all integers in the nested list, including those in sublists.

    Example:
    >>> sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]])
    36
    >>> sum_nested_list([1, [2, 3], [4, [5]]])
    15
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # Check if the element is a list
            total += sum_nested_list(element)  # Recursively sum the nested list
        else:
            total += element  # Add the number to the total
    return total

nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total Sum: ", result)
```
